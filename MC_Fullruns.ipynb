```python
import matplotlib.pyplot as plt

from VectorMC import *
from BlogModel import *

def fullrun(B,steps):
    for attn_norm in ['state','blog']:
        model = VectorSpinMC(B=B,attn_norm=attn_norm,reduction_type='AC')
        m_states = model.long_run(steps)
        model.plot_TEs0(m_states, f"self,{attn_norm}")
        np.save(path+f"fullrun_{attn_norm}_{str(B).replace(".","")}.npy",m_states)

def ness_comparison(B):
    plt.figure(figsize=(4,2))
    for attn_norm in ['state','blog']:
        model = VectorSpinMC(B=B,attn_norm=attn_norm)
        m1 = np.load(path+f"fullrun_{attn_norm}_{str(B).replace(".","")}.npy")
        ness = model.ness_sim(m1)
        plt.plot(ness,label=attn_norm)
    ymax = ness[-1]+0.1
    plt.title(f'B={B}')
    plt.ylim(-1,ymax)
    plt.legend()
    plt.show()
```

```python
fullrun(1000,2)
ness_comparison(1000)
```

```python
fullrun(100,5)
fullrun(10,20)
fullrun(0.1,50)
fullrun(0.01,50)
fullrun(0.001,50)
```

```python
for B in [1000,100,10,1,0.1,0.01,0.001]:
    ness_comparison(B)
```
